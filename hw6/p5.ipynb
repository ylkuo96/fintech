```python
d = 922037
binarystr = bin(d)[2:]
strLen = len(binarystr)

num = 0
cnt = 0
action = []
```

```python
for i in range(strLen):
    if i == 0 and binarystr[i] == '1':
        num = 1
        
        action.append('init')
    elif binarystr[i] == '0':
        num = num * 2

        action.append('d')
    elif binarystr[i] == '1':
        cnt = cnt + 1
        if i == strLen - 1 or binarystr[i+1] == '0':
            if cnt > 2:
                # the faster way
                num = num + 1
                num = num * pow(2, cnt)
                num = num - 1
                
                action.append('a')
                for j in range(cnt):
                    action.append('d')
                action.append('s')                
                
                cnt = 0
            elif cnt == 2 or cnt == 1:
                # the original way (problem 4)
                for j in range(cnt):
                    num = num * 2
                    num = num + 1

                    action.append('d')
                    action.append('a')      

                cnt = 0
```

```python
#print(num)
print('double:', action.count('d'))
print('add:', action.count('a'))
print('subtract:', action.count('s'))
#print(action)
```
